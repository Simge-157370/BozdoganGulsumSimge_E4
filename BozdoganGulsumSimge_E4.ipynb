```python
def nearist_neighbor_function(A, z):
    return A[np.argmin(np.abs(A-z))]
A = np.random.rand(500)
print(f"A: {A}")
u = 1
print("Nearist neighbor: ")
print(nearist_neighbor_function(A, u))
```
